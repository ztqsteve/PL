```python
writer = pd.ExcelWriter('MN_corr_2015.xlsx')
def savecorr(df,category):    
    df_corr = df.corr()[['GROSS SALES']]
    df_corr = df_corr.iloc[:-1,:]
    df_corr = df_corr.sort_values(by='GROSS SALES', ascending = False,)
    df_corr.rename(columns={'GROSS SALES': 'Correlation'}, inplace=True)
    df_corr.to_excel(writer,category)
```

```python
def pltcorrlargehm(k, data, target_var ):
    
    rcParams['figure.figsize'] = 10,10
    corrmat = data.corr()
    cols = corrmat.nlargest(k, target_var)[target_var].index
    cm = np.corrcoef(data[cols].values.T)
    sns.set(font_scale = 1.25)
    hm= sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size':10},
                    yticklabels=cols.values,xticklabels=cols.values)
    plt.show()
```

```python
# filter zip of MN
def filterbyzip(df):
    df['ZIP'] = df['ZIP'].astype(str)
    df['ZIP'] = df['ZIP'].apply(lambda x: x.zfill(5))
    return (df[df['ZIP'].str.startswith('55') | df['ZIP'].str.startswith('56')])

def get_county(df):
    df = df.merge(df_state_zip_map[['Level 3', 'Level 4']],how = 'left', left_on = 'ZIP', right_on = 'Level 4')
    df = df.drop(columns = 'Level 4')
    df['Level 3'] = df['Level 3'].str.rstrip(', MN')
    df['Level 3'] = df['Level 3'].str.rstrip('County')
    df['Level 3'] = df['Level 3'].str.upper()
    df['Level 3'] = df['Level 3'].str.rstrip(' ')
    df = df.drop(['State','ZIP'],axis=1)
    df = df.rename(columns={'Level 3': 'COUNTY'})
    return (df)

def ZipCountyMatch(df):
    df = filterbyzip(df)
    df = get_county(df)
    return (df)
```

```python
df_hh_income = filterbyzip(df_hh_income)
```

```python
df_hh_income = get_county(df_hh_income)
```

```python
df_hh_income_1 = df_hh_income.iloc[:,:-9]
shape_hh_income = df_hh_income_1.shape
for i in range(4,shape_hh_income[1]):
    if i%4 == 0:
        df_hh_income_1['#' + df_hh_income_1.columns[i]] = df_hh_income_1.iloc[:,i].multiply(df_hh_income_1.iloc[:,0])/100
    elif i%4 == 1:
        df_hh_income_1['#' + df_hh_income_1.columns[i]] = df_hh_income_1.iloc[:,i].multiply(df_hh_income_1.iloc[:,1])/100
    elif i%4 == 2:
        df_hh_income_1['#' + df_hh_income_1.columns[i]] = df_hh_income_1.iloc[:,i].multiply(df_hh_income_1.iloc[:,2])/100
    else:
        df_hh_income_1['#' + df_hh_income_1.columns[i]] = df_hh_income_1.iloc[:,i].multiply(df_hh_income_1.iloc[:,3])/100
```

```python
df_hh_income_1['COUNTY'] = df_hh_income['COUNTY']
```

```python
df_hh_income_1_by_county = df_hh_income_1.groupby('COUNTY').sum()
```

```python
def mode1(df,num):
    # one-one matching
    shape = df.shape
    for i in range(num,shape[1]-1,num):
        for j in range(num):
            df['#' + df.columns[i+j]] = df.iloc[:,i+j].multiply(df.iloc[:,j])/100
            
    df_by_county = df.groupby('COUNTY').sum()
    
    for i in range(num,shape[1]-1,num):
        for j in range(num):
            df_by_county[df_by_county.columns[i+j]] = df_by_county.iloc[:,i+j+shape[1]-1-num]/\
            df_by_county.iloc[:,j]*100
   
    return (df_by_county.iloc[:,:shape[1]-1])
```

```python
def mode2(df,num):
    #one-many matching
    shape = df.shape[1]-1
    total_ix = []
    percent_ix = []
    for i in range(shape):
        if i%num == 0:
            total_ix.append(i)
        else:
            percent_ix.append(i)

    for i in range(0,shape,num):
        for j in range(1,num):
            df['#' + df.columns[i+j]] = df.iloc[:,i+j].multiply(df.iloc[:,i])/100

    df_by_county = df.groupby('COUNTY').sum()
    df_by_county_num = df_by_county.iloc[:,shape:]
    df_by_county_total = df_by_county.iloc[:,total_ix]
    df_by_county_percent = df_by_county.iloc[:,percent_ix]
    df_by_county = pd.concat([df_by_county_total,df_by_county_percent,df_by_county_num], axis=1)

    for i,j in zip(range(df_by_county_total.shape[1]),range(df_by_county_total.shape[1],shape,num-1)):
        for k in range(num-1):
            df_by_county.iloc[:,j+k] = df_by_county.iloc[:,j+k+df_by_county_percent.shape[1]]/\
            df_by_county.iloc[:,i]*100
            
    return (df_by_county.iloc[:,:shape])
```

```python
def preprocessSL(category):
    df = pd.read_excel('150_Scorelogix Indices.xlsx',sheet_name = category,header= 1)
    df = ZipCountyMatch(df)
    return (df)
def processSL(df):
    df = pd.concat([df.filter(regex='2015'),df['COUNTY']],axis=1)
    df_by_county = df.groupby('COUNTY').mean()
    df_by_county = df_by_county.reset_index()
    df_rev = df_by_county.merge(df_total_rev_by_county, how = 'inner', on = 'COUNTY')
    return (df_rev,df_by_county)
```
